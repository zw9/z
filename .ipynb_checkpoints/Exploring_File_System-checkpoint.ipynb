```python
import folderstats
dfstats=""
dfstats = folderstats.folderstats('.', ignore_hidden=True )
display(dfstats.tail(10))
```

```python
import os
for (dirpath, dirnames, filenames) in os.walk('.'):
    for f in filenames:
        print('FILE :', os.path.join(dirpath, f))
    for d in dirnames:
        print('DIRECTORY :', os.path.join(dirpath, d))
```
